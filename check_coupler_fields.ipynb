```python
%matplotlib inline
import matplotlib.pyplot as plt
import xarray as xr
```

```python
# filename = 'vic5.RI.test.20161016a.cpl.ha.1948-09.nc'
filename = 'vic5.RI.test.20170212a.cpl.ha.1948-09.nc'
ds5 = xr.open_dataset(filename)
# filename = 'vic4.RI.test.20161117a.cpl.ha.1949-09.nc'
filename = 'vic4.RI.test.20170212a.cpl.ha.1948-09.nc'
ds4 = xr.open_dataset(filename)
```

```python
variable = 'l2xavg_Fall_lwup'

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

ds5[variable].plot(cmap='viridis', ax=ax1, vmax=-210, vmin=-450)
ax1.set_title('VIC5')

ds4[variable].plot(cmap='viridis', ax=ax2, vmax=-210, vmin=-450)
ax2.set_title('VIC4')
```
