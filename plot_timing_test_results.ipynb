```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import os
import matplotlib.dates as mdates
```

```python
cwd = os.getcwd()
ri_reg = xr.open_dataset(os.path.join(cwd, 'timing', 'timing.1hr_1948-09-01-00000.nc'))
ri_20 = xr.open_dataset(os.path.join(cwd, 'timing', 'timing.20min_1948-09-01-00000.nc'))
ri_40nodes = xr.open_dataset(os.path.join(cwd, 'timing_test_results', 'timing_40nodes.nc'))
ri_30nodes = xr.open_dataset(os.path.join(cwd, 'timing_test_results', 'timing_30nodes.nc'))
ri_20nodes = xr.open_dataset(os.path.join(cwd, 'timing_test_results', 'timing_20nodes.nc'))
ri_10nodes = xr.open_dataset(os.path.join(cwd, 'timing_test_results', 'timing_10nodes.nc'))
```

```python
vars = ['OUT_TIME_VICRUN_WALL', 'OUT_TIME_VICRUN_CPU', 'OUT_SNOWT_FBFLAG', 'OUT_SURFT_FBFLAG', 
        'OUT_TCAN_FBFLAG', 'OUT_TFOL_FBFLAG']
```

Max run time, Compare number of soil nodes (colorbars *not* constrained)

```python
fig, axes = plt.subplots(1,5, figsize=(25,4))
var = 'OUT_TIME_VICRUN_WALL'
kwargs={'label': 'Max VICRUN Wall Time (s)'}
(ri_reg[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[0], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              )
axes[0].set_title("50 Nodes")

(ri_40nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[1], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              )
axes[1].set_title("40 Nodes")

(ri_30nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[2], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              )
axes[2].set_title("30 Nodes")

(ri_20nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[3], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              )
axes[3].set_title("20 Nodes")

(ri_10nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[4], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              cbar_kwargs=kwargs)
axes[4].set_title("10 Nodes")

direc = "/Users/diana/Dropbox/UW/Research/rasm/compare_coupler_hist_files/RBR_hist_files/timing_plots"
plotname = "timing_nodes_comp_cbar_notconstrained.png"
savepath = os.path.join(direc, plotname)
plt.savefig(savepath, dpi=100, bbox_inches='tight')
```

Max run time, Compare number of soil nodes (colorbars *are* constrained)

```python
fig, axes = plt.subplots(1,5, figsize=(25,4))
var = 'OUT_TIME_VICRUN_WALL'
kwargs={'label': 'Max VICRUN Wall Time (s)'}
(ri_reg[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[0], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              robust=True)
axes[0].set_title("50 Nodes")

(ri_40nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[1], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              robust=True)
axes[1].set_title("40 Nodes")

(ri_30nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[2], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              robust=True)
axes[2].set_title("30 Nodes")

(ri_20nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[3], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              robust=True)
axes[3].set_title("20 Nodes")

(ri_10nodes[var]/np.timedelta64(1,'s')).max(axis=0).plot(ax=axes[4], 
                                              cmap='viridis', 
                                              add_labels=False, 
                                              robust=True, 
                                              cbar_kwargs=kwargs)
axes[4].set_title("10 Nodes")

direc = "/Users/diana/Dropbox/UW/Research/rasm/compare_coupler_hist_files/RBR_hist_files/timing_plots"
plotname = "timing_nodes_comp_cbar_constrained.png"
savepath = os.path.join(direc, plotname)
plt.savefig(savepath, dpi=100, bbox_inches='tight')
```
