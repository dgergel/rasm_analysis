```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
import xarray as xr
```

```python
direc = '/Users/diana/Dropbox/UW/Research/rasm/25_km'
ncparams_50km = 'vic_params_joe_copy.nc'
params_50km = xr.open_dataset(os.path.join(direc, ncparams_50km))
```

look at total soil depth

```python
depths_sum = params_50km.depth.isel(nlayer=0) + params_50km.depth.isel(nlayer=1) + params_50km.depth.isel(nlayer=2)
depths_sum.plot(cmap='viridis', 
                add_labels=False, 
                cbar_kwargs={'label':'Total Soil Depth (m)'},
                )
```

```python
fig, axes = plt.subplots(1,3, figsize=(16,4))
kwargs={'cbar_label': 'VIC Run Time (s)'}
params_50km.depth.isel(nlayer=0).plot(cmap='viridis', 
                                      ax=axes[0],
                                      add_labels=False, 
                                      cbar_kwargs={'label':'Soil Depth (m)'},
                                      vmin=0.29,
                                      vmax=0.31,
                                      )
axes[0].set_title("First Layer")

params_50km.depth.isel(nlayer=1).plot(cmap='viridis', 
                                      ax=axes[1],
                                      add_labels=False, 
                                      robust=True, 
                                      cbar_kwargs={'label':'Soil Depth (m)'},
                                      )
axes[1].set_title("Second Layer")

params_50km.depth.isel(nlayer=2).plot(cmap='viridis', 
                                      ax=axes[2],
                                      add_labels=False, 
                                      robust=True, 
                                      cbar_kwargs={'label':'Soil Depth (m)'},
                                      )
axes[2].set_title("Third Layer")
```
