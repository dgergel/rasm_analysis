```python
%matplotlib inline
import xarray as xr
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
direc = '/u/home/gergel/rasm_joe_landsurface_paper/R1009RBRceap01a/lnd/monthly_mean_timeseries'
filename = 'R1009RBRceap01a.vic.hmm.197909-201412.nc'
monthly_means = xr.open_dataset(os.path.join(direc, filename))
```

```python
monthly_means.Soilw
```

```python
prec_grps = monthly_means.Soilw.isel(depth=0).groupby('time.month').mean('time')
fig, axes = plt.subplots(2, 6, figsize=(22, 6))

prec_grps[0, :, :].where(prec_grps[0, :, :] < 2).plot(robust=True, ax=axes[0, 0], add_labels=False)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

'''plt.setp(axes[0, 0].get_xticklabels(), visible=False)
plt.setp(axes[0, 0].get_yticklabels(), visible=False)'''
prec_grps[1, :, :].where(prec_grps[1, :, :] < 2).plot(robust=True, ax=axes[0, 1], add_labels=False)
prec_grps[2, :, :].where(prec_grps[2, :, :] < 2).plot(robust=True, ax=axes[0, 2], add_labels=False)
prec_grps[3, :, :].where(prec_grps[3, :, :] < 2).plot(robust=True, ax=axes[0, 3], add_labels=False)
prec_grps[4, :, :].where(prec_grps[4, :, :] < 2).plot(robust=True, ax=axes[0, 4], add_labels=False)
prec_grps[5, :, :].where(prec_grps[5, :, :] < 2).plot(robust=True, ax=axes[0, 5], add_labels=False, cbar_kwargs={'label':'Fractional Soil Moisture'})
prec_grps[6, :, :].where(prec_grps[6, :, :] < 2).plot(robust=True, ax=axes[1, 0], add_labels=False)
prec_grps[7, :, :].where(prec_grps[7, :, :] < 2).plot(robust=True, ax=axes[1, 1], add_labels=False)
prec_grps[8, :, :].where(prec_grps[8, :, :] < 2).plot(robust=True, ax=axes[1, 2], add_labels=False)
prec_grps[9, :, :].where(prec_grps[9, :, :] < 2).plot(robust=True, ax=axes[1, 3], add_labels=False)
prec_grps[10, :, :].where(prec_grps[10, :, :] < 2).plot(robust=True, ax=axes[1, 4], add_labels=False)
prec_grps[11, :, :].where(prec_grps[11, :, :] < 2).plot(robust=True, ax=axes[1, 5], add_labels=False, cbar_kwargs={'label':'Fractional Soil Moisture'})

count = 0
for i in np.arange(0, 2): 
    for j in np.arange(0, 6):
        plt.setp(axes[i, j].get_xticklabels(), visible=False)
        plt.setp(axes[i, j].get_yticklabels(), visible=False)
        axes[i, j].set_title(months[count])
        count += 1
plt.suptitle('First Layer Monthly Mean Soil Moisture, Averaged over 1979-2014, RBR run from Joe\'s paper')
```
