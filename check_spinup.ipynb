```python
%matplotlib inline
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import os
```

```python
workdir = os.getcwd()
monthly_filename = 'fluxes_monthly4.1978-01.nc'
mo_fluxes = xr.open_dataset(os.path.join(workdir, monthly_filename))
```

```python
mo_fluxes.OUT_SOIL_TNODE
```

```python
mo_fluxes.OUT_SOIL_TNODE[:, :, 73, 130].sel(node=2).plot()
ts0 = mo_fluxes.OUT_SOIL_TNODE[:, :, 73, 130].sel(node=2)
m, b = np.polyfit(np.arange(0, len(mo_fluxes.time)), ts0, 1)
plt.plot(mo_fluxes.time, m*ts0 + b, '--', color='k', linewidth=3.0)
print(m)
print(b)
```
