```python
fig, axes = plt.subplots(1,2, figsize=(10,4))
var = 'l2x_Sl_logz0'
vmin=-1.5
vmax=1.5
vic4[var].plot(ax=axes[0], cmap='viridis', add_labels=False, robust=True, vmin=vmin, vmax=vmax)
axes[0].set_title("VIC 4")
vic5[var].plot(ax=axes[1], cmap='viridis', add_labels=False, robust=True, vmin=vmin, vmax=vmax)
axes[1].set_title("VIC 5")
```

compare RI `l2x_Fall_lat` with `OUT_LATENT` from image driver (constrain colorbars)

```python
fig, axes = plt.subplots(1,3, figsize=(13,4))
vmin=0
vmax=240

vic4_ri_3rd['l2x_Fall_lat'].plot(ax=axes[0], cmap='viridis',
                                 add_labels=False,
                                 vmin=vmin, vmax=vmax)
axes[0].set_title('VIC4 RI')

vic5_ri_3rd['l2x_Fall_lat'].plot(ax=axes[1], cmap='viridis',
                                 add_labels=False,
                                 vmin=vmin, vmax=vmax)
axes[1].set_title('VIC5 RI')

fluxes['OUT_LATENT'].isel(time=1).plot(ax=axes[2], cmap='viridis',
                                       add_labels=False,
                                       vmin=vmin, vmax=vmax)
axes[2].set_title('Image Driver')
```

compare RI `l2x_Flrl_rofliq` with the sum of `OUT_RUNOFF` and `OUT_BASEFLOW` from image driver

```python
fig, axes = plt.subplots(1,3, figsize=(13,4))
vmin=0
vmax=0.001

vic4_ri_3rd['l2x_Flrl_rofliq'].plot(ax=axes[0], cmap='viridis',
                                    add_labels=False,
                                    vmin=vmin, vmax=vmax)
axes[0].set_title('VIC4 RI')

vic5_ri_3rd['l2x_Flrl_rofliq'].plot(ax=axes[1], cmap='viridis',
                                    add_labels=False,
                                    vmin=vmin, vmax=vmax)
axes[1].set_title('VIC5 RI')

dt = 3600*3
rofliq = (fluxes['OUT_RUNOFF'].isel(time=0) + fluxes['OUT_BASEFLOW'].isel(time=0)) / dt

rofliq.plot(ax=axes[2], 
            cmap='viridis',
            add_labels=False,
            vmin=vmin, vmax=vmax)
axes[2].set_title('Image Driver')
```
